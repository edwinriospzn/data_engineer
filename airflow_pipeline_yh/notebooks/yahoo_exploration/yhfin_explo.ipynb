```python
import yfinance as yf
import pandas as pd

tickers = ["AAPL"]#, "NVDA"]

df = yf.download(
    tickers=tickers,
    period="1d",
    interval="1m",
    group_by="ticker",
    progress=False
)

# Convert to long format with a ticker column
if isinstance(df.columns, pd.MultiIndex):
    df = df.stack(level=0).rename_axis(index=["Date", "ticker"]).reset_index()

# Optional: keep only the columns you want
# df = df[["Date", "ticker", "Close"]]

print(df.head())
```

```python
import yfinance as yf

# Example: get fundamental data for each ticker
for ticker in ["AAPL", "NVDA"]:
    stock = yf.Ticker(ticker)
    info = stock.info
    print(ticker)
    print("Market Cap:", info.get("marketCap"))
    print("PE Ratio:", info.get("trailingPE"))
    print("Dividend Yield:", info.get("dividendYield"))
    print("Sector:", info.get("sector"))
    print("-" * 40)

# If you want a table, collect them into a DataFrame
fundamentals = []
for ticker in ["AAPL", "NVDA"]:
    stock = yf.Ticker(ticker)
    info = stock.info
    fundamentals.append({
        "ticker": ticker,
        "marketCap": info.get("marketCap"),
        "trailingPE": info.get("trailingPE"),
        "dividendYield": info.get("dividendYield"),
        "sector": info.get("sector")
    })

fundamentals_df = pd.DataFrame(fundamentals)
fundamentals_df
```
