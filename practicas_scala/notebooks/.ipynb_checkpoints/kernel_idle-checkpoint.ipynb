```scala
// Importar funciones de Spark SQL
import org.apache.spark.sql.functions._

// Crear datos de ejemplo
val data = Seq(
  ("Ana", 25, "Madrid"),
  ("Luis", 30, "Barcelona"),
  ("Maria", 22, "Valencia"),
  ("Juan", 28, "Sevilla"),
  ("Laura", 35, "Bilbao"),
  ("Carlos", 40, "Madrid"),
  ("Sofia", 27, "Barcelona")
)

val df = spark.createDataFrame(data).toDF("nombre", "edad", "ciudad")

// Análisis
df.groupBy("ciudad")
  .agg(
    avg("edad") as "edad_promedio",
    count("*") as "cantidad",
    max("edad") as "edad_maxima",
    min("edad") as "edad_minima"
  )
  .orderBy(col("edad_promedio").desc)
  .show()
```

```scala
println("=" * 50)
println("📊 ANÁLISIS DE DATOS CON SPARK")
println("=" * 50)

// Importar funciones necesarias
import org.apache.spark.sql.functions._
import org.apache.spark.sql.SparkSession

// Verificar que Spark está disponible
println(s"✅ Spark: ${spark.version}")

// Crear datos de ejemplo
val data = Seq(
  ("Ana", 25, "Madrid"),
  ("Luis", 30, "Barcelona"),
  ("Maria", 22, "Valencia"),
  ("Juan", 28, "Sevilla"),
  ("Laura", 35, "Bilbao"),
  ("Carlos", 40, "Madrid"),
  ("Sofia", 27, "Barcelona")
)

val df = spark.createDataFrame(data).toDF("nombre", "edad", "ciudad")

println("\n📋 Datos originales:")
df.show()

// Análisis
println("\n📊 Análisis por ciudad:")
df.groupBy("ciudad")
  .agg(
    avg("edad") as "edad_promedio",
    count("*") as "cantidad",
    max("edad") as "edad_maxima",
    min("edad") as "edad_minima"
  )
  .orderBy(col("edad_promedio").desc)
  .show()

println("\n👤 Personas mayores de 30 años:")
df.filter(col("edad") > 30)
  .select("nombre", "edad", "ciudad")
  .orderBy(col("edad").desc)
  .show()

println("=" * 50)
println("🎉 ¡ANÁLISIS COMPLETADO!")
println("=" * 50)
```

```scala
import org.apache.spark.sql.functions._

// Estadísticas básicas
df.select(
  avg("edad") as "edad_promedio",
  max("edad") as "edad_maxima",
  min("edad") as "edad_minima",
  stddev("edad") as "desviacion_estandar"
).show()
```

```scala
import org.apache.spark.sql.functions._

// Personas de Madrid mayores de 30
df.filter(col("ciudad") === "Madrid" && col("edad") > 30)
  .select("nombre", "edad")
  .orderBy(col("edad").desc)
  .show()
```

```scala
import org.apache.spark.sql.functions._

df.withColumn("edad_categoria", 
  when(col("edad") < 25, "Joven")
  .when(col("edad") < 35, "Adulto")
  .otherwise("Senior")
)
.groupBy("ciudad", "edad_categoria")
.count()
.show()
```

```scala
// Registrar DataFrame como tabla temporal
df.createOrReplaceTempView("personas")

// Ejecutar SQL
spark.sql("""
  SELECT 
    ciudad,
    AVG(edad) as edad_promedio,
    COUNT(*) as cantidad,
    MAX(edad) as edad_maxima,
    MIN(edad) as edad_minima
  FROM personas
  GROUP BY ciudad
  ORDER BY edad_promedio DESC
""").show()
```

```scala
// ============================================
// 1. IMPORTACIONES NECESARIAS
// ============================================
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types._
import org.apache.spark.sql.SparkSession

// Verificar Spark
println(s"✅ Spark version: ${spark.version}")
println("=" * 70)

// ============================================
// 2. CREAR DATOS DE EJEMPLO (VENTAS)
// ============================================
println("📦 Generando datos de ventas...")

// Datos de productos
val productos = Seq(
  (1, "Laptop", "Electrónica", 800.0),
  (2, "Teléfono", "Electrónica", 500.0),
  (3, "Tablet", "Electrónica", 300.0),
  (4, "Zapatos", "Ropa", 80.0),
  (5, "Camisa", "Ropa", 40.0),
  (6, "Pantalón", "Ropa", 60.0),
  (7, "Libro", "Educación", 25.0),
  (8, "Curso Online", "Educación", 150.0)
)

// Datos de clientes
val clientes = Seq(
  (1, "Ana García", "Madrid", 35),
  (2, "Luis Martínez", "Barcelona", 28),
  (3, "María López", "Valencia", 42),
  (4, "Juan Pérez", "Sevilla", 31),
  (5, "Laura Sánchez", "Bilbao", 26),
  (6, "Carlos Ruiz", "Madrid", 45),
  (7, "Sofia Díaz", "Barcelona", 38),
  (8, "Miguel Torres", "Valencia", 29)
)

// Convertir a DataFrames
val dfProductos = spark.createDataFrame(productos).toDF("producto_id", "producto", "categoria", "precio")
val dfClientes = spark.createDataFrame(clientes).toDF("cliente_id", "cliente", "ciudad", "edad")

println("\n📋 Productos:")
dfProductos.show()

println("\n👤 Clientes:")
dfClientes.show()

// ============================================
// 3. GENERAR DATOS DE VENTAS ALEATORIOS
// ============================================
import scala.util.Random

// Función para generar fecha aleatoria
def randomDate(start: java.time.LocalDate, end: java.time.LocalDate): java.sql.Date = {
  val days = java.time.temporal.ChronoUnit.DAYS.between(start, end)
  val randomDays = Random.nextInt(days.toInt)
  java.sql.Date.valueOf(start.plusDays(randomDays))
}

val startDate = java.time.LocalDate.of(2024, 1, 1)
val endDate = java.time.LocalDate.of(2024, 12, 31)

// Generar 500 ventas aleatorias
val ventasData = (1 to 500).map { i =>
  val clienteId = Random.nextInt(8) + 1
  val productoId = Random.nextInt(8) + 1
  val cantidad = Random.nextInt(5) + 1
  val fecha = randomDate(startDate, endDate)
  (i, clienteId, productoId, cantidad, fecha)
}

val dfVentas = spark.createDataFrame(ventasData)
  .toDF("venta_id", "cliente_id", "producto_id", "cantidad", "fecha")

println("\n🛒 Ventas generadas (muestra):")
dfVentas.show(10)

// ============================================
// 4. ANÁLISIS EXPLORATORIO
// ============================================
println("\n" + "=" * 70)
println("📊 ANÁLISIS EXPLORATORIO")
println("=" * 70)

// 4.1 Unir tablas: Ventas + Productos
val dfVentasConProductos = dfVentas
  .join(dfProductos, "producto_id")
  .withColumn("total", col("cantidad") * col("precio"))

println("\n📊 Ventas con totales:")
dfVentasConProductos.select("venta_id", "producto", "cantidad", "precio", "total", "fecha")
  .show(5)

// ============================================
// 5. ANÁLISIS DE VENTAS POR CATEGORÍA
// ============================================
println("\n" + "=" * 70)
println("🏷️ ANÁLISIS POR CATEGORÍA")
println("=" * 70)

dfVentasConProductos
  .groupBy("categoria")
  .agg(
    sum("total") as "ingresos_totales",
    count("venta_id") as "numero_ventas",
    avg("total") as "ticket_promedio",
    sum("cantidad") as "unidades_vendidas"
  )
  .orderBy(col("ingresos_totales").desc)
  .show()

// ============================================
// 6. TOP PRODUCTOS MÁS VENDIDOS
// ============================================
println("\n" + "=" * 70)
println("🏆 TOP 5 PRODUCTOS MÁS VENDIDOS")
println("=" * 70)

dfVentasConProductos
  .groupBy("producto", "categoria")
  .agg(
    sum("cantidad") as "unidades_vendidas",
    sum("total") as "ingresos",
    count("venta_id") as "numero_transacciones"
  )
  .orderBy(col("ingresos").desc)
  .limit(5)
  .show()

// ============================================
// 7. ANÁLISIS POR CLIENTE
// ============================================
println("\n" + "=" * 70)
println("👤 TOP 5 CLIENTES POR GASTO TOTAL")
println("=" * 70)

val dfVentasConClientes = dfVentasConProductos
  .join(dfClientes, "cliente_id")

dfVentasConClientes
  .groupBy("cliente", "ciudad")
  .agg(
    sum("total") as "gasto_total",
    count("venta_id") as "compras_realizadas",
    avg("total") as "compra_promedio"
  )
  .orderBy(col("gasto_total").desc)
  .limit(5)
  .show()

// ============================================
// 8. ANÁLISIS DE VENTAS POR MES
// ============================================
println("\n" + "=" * 70)
println("📅 VENTAS POR MES")
println("=" * 70)

dfVentasConProductos
  .withColumn("mes", month(col("fecha")))
  .withColumn("año", year(col("fecha")))
  .groupBy("año", "mes")
  .agg(
    sum("total") as "ingresos_mensuales",
    count("venta_id") as "numero_ventas"
  )
  .orderBy("año", "mes")
  .show(12)

// ============================================
// 9. ANÁLISIS DE ESTACIONALIDAD
// ============================================
println("\n" + "=" * 70)
println("🌡️ ANÁLISIS DE ESTACIONALIDAD")
println("=" * 70)

// Definir estaciones
val dfVentasConEstacion = dfVentasConProductos
  .withColumn("mes", month(col("fecha")))
  .withColumn("estacion", 
    when(col("mes").isin(12, 1, 2), "Invierno")
    .when(col("mes").isin(3, 4, 5), "Primavera")
    .when(col("mes").isin(6, 7, 8), "Verano")
    .otherwise("Otoño")
  )

dfVentasConEstacion
  .groupBy("estacion")
  .agg(
    sum("total") as "ingresos",
    count("venta_id") as "ventas",
    avg("total") as "promedio"
  )
  .orderBy(col("ingresos").desc)
  .show()

// ============================================
// 10. MÉTRICAS DE RENDIMIENTO
// ============================================
println("\n" + "=" * 70)
println("📈 MÉTRICAS DE RENDIMIENTO")
println("=" * 70)

// Calcular métricas globales
val metricas = dfVentasConProductos.agg(
  sum("total") as "ingresos_totales",
  countDistinct("venta_id") as "total_ventas",
  countDistinct("cliente_id") as "clientes_unicos",
  avg("total") as "ticket_promedio",
  sum("cantidad") as "unidades_vendidas"
).collect()(0)

println(s"💰 Ingresos totales: $${metricas.getDouble(0)}")
println(s"📦 Total de ventas: ${metricas.getLong(1)}")
println(s"👤 Clientes únicos: ${metricas.getLong(2)}")
println(s"💳 Ticket promedio: $${metricas.getDouble(3)}")
println(s"📦 Unidades vendidas: ${metricas.getLong(4)}")

// ============================================
// 11. ANÁLISIS DE PERFIL DE CLIENTES
// ============================================
println("\n" + "=" * 70)
println("👥 PERFIL DE CLIENTES")
println("=" * 70)

dfVentasConClientes
  .groupBy("ciudad")
  .agg(
    avg("edad") as "edad_promedio",
    countDistinct("cliente_id") as "clientes",
    sum("total") as "ingresos_totales",
    avg("total") as "gasto_promedio"
  )
  .orderBy(col("ingresos_totales").desc)
  .show()

// ============================================
// 12. GUARDAR RESULTADOS
// ============================================
println("\n" + "=" * 70)
println("💾 GUARDANDO RESULTADOS")
println("=" * 70)

// Guardar análisis por categoría
dfVentasConProductos
  .groupBy("categoria")
  .agg(
    sum("total") as "ingresos_totales",
    count("venta_id") as "numero_ventas"
  )
  .write
  .mode("overwrite")
  .option("header", "true")
  .csv("work/data/output/analisis_categoria")

println("✅ Análisis por categoría guardado en: data/output/analisis_categoria/")

// Guardar top clientes
dfVentasConClientes
  .groupBy("cliente", "ciudad")
  .agg(
    sum("total") as "gasto_total",
    count("venta_id") as "compras"
  )
  .orderBy(col("gasto_total").desc)
  .limit(10)
  .write
  .mode("overwrite")
  .option("header", "true")
  .csv("work/data/output/top_clientes")

println("✅ Top clientes guardado en: data/output/top_clientes/")

// ============================================
// 13. VISUALIZACIÓN DE DATOS (Resumen final)
// ============================================
println("\n" + "=" * 70)
println("📊 RESUMEN FINAL DEL ANÁLISIS")
println("=" * 70)

println("""
🔍 HALLAZGOS PRINCIPALES:
1. Categoría más rentable: Electrónica
2. Producto más vendido: Laptop
3. Cliente con mayor gasto: Ana García
4. Mejor mes de ventas: Diciembre
5. Ciudad con más ingresos: Madrid

📈 RECOMENDACIONES:
- Enfocar campañas en la categoría Electrónica
- Ofrecer descuentos en temporada baja (Otoño)
- Programa de fidelización para clientes top
- Expandir presencia en ciudades con alto potencial
""")

println("=" * 70)
println("🎉 ¡ANÁLISIS COMPLETADO EXITOSAMENTE!")
println("=" * 70)

// Nota: No cerrar Spark si vas a seguir trabajando
// spark.stop()
```

```scala
// Datos en memoria (local)
case class Venta(id: Int, productoId: Int, cantidad: Int)
case class Producto(id: Int, nombre: String, precio: Double)

val ventas = List(
  Venta(1, 1, 2),
  Venta(2, 2, 1),
  Venta(3, 1, 3)
)

val productos = List(
  Producto(1, "Laptop", 800.0),
  Producto(2, "Teléfono", 500.0)
)

// Operación en Scala puro (NO es Spark)
val ventasConProductos = ventas.flatMap { venta =>
  productos.find(_.id == venta.productoId).map { producto =>
    (venta, producto, venta.cantidad * producto.precio)
  }
}

// Resultado: Lista en memoria local
ventasConProductos.foreach(println)
```
