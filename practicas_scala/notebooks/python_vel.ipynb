```python
# ============================================
# PYTHON - Prueba de velocidad equivalente
# ============================================
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import time


```

```python
# Crear SparkSession
spark = SparkSession.builder \
    .appName("Python-Test") \
    .config("spark.sql.adaptive.enabled", "false") \
    .getOrCreate()
# Medir tiempo
start = time.time()

# Procesar datos
total = spark.range(10000000) \
    .withColumn("cuadrado", col("id") * col("id")) \
    .filter(col("cuadrado") > 1000000) \
    .count()

# Calcular tiempo
end = time.time()
tiempo = end - start

# Mostrar resultados
print(f"🐍 Python: {total} registros encontrados")
print(f"⏱️ Tiempo: {tiempo:.3f} segundos")

# Cerrar Spark
spark.stop()
```
